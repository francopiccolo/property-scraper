```python
import pandas as pd
pd.set_option('display.max_colwidth', None)

df = pd.read_csv('../data/preprocessed/terrenos/df2.csv')
```

```python
price_mask = df['price'] > 5000
price_per_square_meter_mask = df['price_per_square_meter'] > 15

df[price_mask & price_per_square_meter_mask].sort_values(by='price_per_square_meter')[:40]
```

```python
import gmaps
import gmaps.datasets
import sys
sys.path.append("../")
from src.config import GMAPS_API_KEY

# Use google maps api key
gmaps.configure(api_key=GMAPS_API_KEY)

locations = []
for property in df.itertuples():
    locations.append((property.lat, property.long))

# Create the map
fig = gmaps.figure(
    center=(-32.934552, -60.666805), 
    zoom_level=10,    
)

heatmap_layer = gmaps.heatmap_layer(
    locations, 
    point_radius=20,
    # dissipating not working
)

fig.add_layer(heatmap_layer)

symbol_layer = gmaps.symbol_layer(
    locations,
    scale=2
)

fig.add_layer(symbol_layer)

fig
```
